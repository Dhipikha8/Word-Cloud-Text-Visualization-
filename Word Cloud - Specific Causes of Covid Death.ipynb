```python
import pandas as pd #Data analysis and manipulation
import matplotlib.pyplot as plt #Data visualization
from wordcloud import WordCloud
```

```python
df= pd.read_csv("covid+death.csv")
```

```python
df.columns
```

```python
df.groupby(["Condition"]).count()
```

```python
df.groupby(["Condition Group"]).count()
```

```python
sentence=df["Condition"].tolist()
```

```python
string=' '.join(sentence)
```

```python
plt.figure(figsize=(14,14))
plt.imshow(WordCloud().generate(string))
```

```python
sentences=df["Condition Group"].tolist()
```

```python
strings=' '.join(sentences)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(WordCloud().generate(strings))
```
